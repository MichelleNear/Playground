```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#to make random data say x = random.sample(range(1, 10), 5) this does a range between 1-10 making 5 random data points
rose_x = [1, 3, 4, 5, 5]
rose_y = [2, 2, 3, 4, 5]
violet_x = [6, 8, 8, 9, 10]
violet_y = [6, 7, 9, 9, 9]
```

```python
plt.scatter(rose_x, rose_y, color = 'deeppink', label = "Rose")
plt.scatter(violet_x, violet_y, color = 'purple', label = "Violet")
plt.xlabel("Width")
plt.ylabel("Length")
plt.title("Roses vs Violets", color = 'green')
plt.legend(loc='upper left')
```

```python
#Linear regression
xr = np.array(rose_x)
yr = np.array(rose_y)
mr, br= np.polyfit(xr, yr, 1)
print(mr)
print(br)
plt.plot(xr, yr, 'o', color = 'deeppink')
plt.plot(xr, mr*xr + br, color = 'darkolivegreen')

x = np.array(violet_x)
y = np.array(violet_y)
m, b = np.polyfit(x, y, 1)
print(m)
print(b)
plt.plot(x, y, 'o', color = 'purple')
plt.plot(x, m*x + b, color = 'yellowgreen')

plt.title("Linear regression", color = 'teal')
```

```python
rose_x = np.array([1, 3, 4, 5, 5]).reshape((-1,1))
rose_y = [2, 2, 3, 4, 5]

model =  LinearRegression()
model.fit(rose_x, rose_y)
b0 = model.intercept_
b1 = model.coef_
plt.scatter(rose_x, rose_y)
plt.plot(rose_x, b1*rose_x + b0)
```
