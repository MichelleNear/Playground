```python
'''MACHINE LEARNING START'''
import matplotlib.pyplot as plt
import numpy as np
from random import randint
```

```python
class flower:
    def get_l(self):
        return self.length
    
    def get_w(self):
        return self.width
    
    def get_n(self):
        return self.name
    
    def set_l(self, val):
        self.length = val
        
    def set_w(self,val):
        self.width = val

class roses(flower):
    def __init__(self):
        self.length = randint(0,3)
        self.width = randint(0,3)
        self.name = 'rose'
    
class daisy(flower):
    def __init__(self):
        self.length = randint(4,5)
        self.width = randint(4,5)
        self.name = 'daisy'
    
#class sunflower(flower):
```

```python
#sum of square error
def sse(guess,truth):
    error = 0
    n = len(guess)
    for i in range(n):
        if guess[i] != truth[i]:
            error += 1**2
    return error/n
```

```python
roseSet = (roses(),roses(),roses())
daisySet = (daisy(),daisy(),daisy())
```

```python
numberofpoints = 6

xflowers = np.zeros(numberofpoints) #Makes array of 6 0s
yflowers = np.zeros(numberofpoints)
truthFlowers = ['a','a','a','a','a','a']
for i in range(int(numberofpoints/2)):
    xflowers[i]=roseSet[i].get_l() #i = 0 [ME,-,-,-,-,-] i = 1 [-,ME,-,-,-,-] i = 2 [-,-,ME,-,-,-]
    xflowers[i+3]=daisySet[i].get_l() #i = 0 [-,-,-,ME,-,-] i = 1 [-,-,-,-,ME,-] i = 2 [-,-,-,-,-,ME]
    
    yflowers[i]=roseSet[i].get_w()
    yflowers[i+3]=daisySet[i].get_w()
    
    truthFlowers[i]=roseSet[i].get_n()
    truthFlowers[i+3]=daisySet[i].get_n()

#Init vals

def makeModel(m,b):
    mymodelY = np.zeros(numberofpoints)
    mymodelX = np.zeros(numberofpoints)
    for i in range(numberofpoints):
        mymodelY[i] = m*i+b
        mymodelX[i] = i
    return mymodelY, mymodelX

```

```python
#evaluate
#What was our guess?
def guess(yflowers,mymodelY):
    guessFlowers = []
    for i in range(numberofpoints):
        if yflowers[i] < mymodelY[i]:
            guessFlowers.append('roses')
        else :
            guessFlowers.append('daisy')
    return guessFlowers
fig, ax = plt.subplots(figsize=(12, 6))

##LEARN!
max_iterations = 1000
mBest = 1
bBest = 0
mymodelY, mymodelX = makeModel(mBest,bBest)
print(mymodelY)
guessFlowers = guess(yflowers,mymodelY)
print(guessFlowers)
sseBest = sse(guessFlowers,truthFlowers)
fig, ax = plt.subplots(figsize=(12, 6))


for i in range(3):
    m = randint(-10,10)
    b = randint(0,7)
    
    mymodelY, null = makeModel(m,b)
    plt.scatter(xflowers, yflowers)
    plt.plot(mymodelX, mymodelY)
    ax.plot((mymodelX), color='blue', label='Sine wave')
    ax.plot((mymodelX), color='black', label='Cosine wave')
    plt.show()
    ssetmp = sse(mymodelY,truthFlowers)
        
    if ssetmp > sseBest:
        mBest= m 
        #print(ssetmp)
```
